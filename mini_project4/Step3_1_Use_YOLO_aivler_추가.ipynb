- 이름 바꾸기

```python
def rename_files(directory, old_prefix, new_prefix, extension=None):
    for filename in os.listdir(directory):
        # 파일명이 old_prefix로 시작하는지 확인
        if filename.startswith(old_prefix):
            # 특정 확장자로 필터링할 경우
            if extension:
                if not filename.endswith(extension):
                    continue
            # 새로운 파일명 생성
            new_filename = filename.replace(old_prefix, new_prefix, 1)
            src = os.path.join(directory, filename)
            dst = os.path.join(directory, new_filename)

            # 파일명 중복 방지
            if os.path.exists(dst):
                print(f'경고: {dst} 파일이 이미 존재합니다. {filename}을(를) 건너뜁니다.')
                continue
            try:
                os.rename(src, dst)
                print(f'파일 이름 변경: {filename} -> {new_filename}')
            except Exception as e:
                print(f'파일 {filename} 변경 중 오류 발생: {e}')

# 디렉토리 경로 설정
images_dir = '/content/member/seo/images'
labels_dir = '/content/member/seo/labels'

# 파일명 이름_face_숫자.jpg .txt 변경
rename_files(images_dir, 'my_face_', 'seo_face_', extension='.jpg')
rename_files(labels_dir, 'my_face_', 'seo_face_', extension='.txt')
```

#### 3) 각 폴더에 데이터 옮기기

- 내 얼굴

```python
image_path = '/content/my_image/my_image/images'
label_path = '/content//my_image/my_image/labels'

all_images = [f for f in os.listdir(image_path) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]
num_total = len(all_images)

# 분할 비율 0.2 0.1에 따라 이미지 몇장인지 계산
test_split = int(0.2 * num_total)
val_split = int(0.1 * num_total)
train_split = num_total - test_split - val_split
```

```python
## 얼굴 데이터를 Training set, Test set으로 분할하기 위한 사전 작업
## 분할 재현성을 위한 난수 2024 고정
random.seed(2024)
random.shuffle(all_images)

all_images[:5]
```

```python
for idx, image_file in enumerate(all_images):
    # 이미지와 라벨 폴더 경로
    src_image_path = os.path.join(image_path, image_file)
    label_file = os.path.splitext(image_file)[0] + '.txt'
    src_label_path = os.path.join(label_path, label_file)

    # train test val 에 따라 목적지 폴더 결정
    if idx < train_split:
        dest_image_path = os.path.join(train_i, image_file)
        dest_label_path = os.path.join(train_l, label_file)
    elif idx < train_split + val_split:
        dest_image_path = os.path.join(val_i, image_file)
        dest_label_path = os.path.join(val_l, label_file)
    else:
        dest_image_path = os.path.join(test_i, image_file)
        dest_label_path = os.path.join(test_l, label_file)

    # 이미지 파일
    shutil.copy(src_image_path, dest_image_path)

    # 라벨 파일이 있는 경우 동일 데이터셋으로 복사
    if os.path.exists(src_label_path):
        shutil.copy(src_label_path, dest_label_path)

print("이미지와 라벨 파일이 train, val, test 폴더에 분할되어 저장되었습니다.")
```

```python
def copy_files(base_path):
    for split in ['test', 'train', 'valid']:  # 원본 파일에 맞춰 'valid' 사용
        # 이미지와 라벨 폴더 경로
        images_path = os.path.join(base_path, split, 'images')
        labels_path = os.path.join(base_path, split, 'labels')

        # 목적지 폴더의 split을 'valid' 대신 'val'로 매핑
        if split == 'test':
            dest_image_path = test_i
            dest_label_path = test_l
        elif split == 'train':
            dest_image_path = train_i
            dest_label_path = train_l
        elif split == 'valid':
            dest_image_path = val_i
            dest_label_path = val_l

        # 이미지 파일 복사
        for img_file in os.listdir(images_path):
            src_img_path = os.path.join(images_path, img_file)
            dst_img_path = os.path.join(dest_image_path, img_file)
            shutil.copy(src_img_path, dst_img_path)

        # 라벨 파일 복사
        for label_file in os.listdir(labels_path):
            src_label_path = os.path.join(labels_path, label_file)
            dst_label_path = os.path.join(dest_label_path, label_file)
            shutil.copy(src_label_path, dst_label_path)
```

```python
image_path = '/content/member/seo/images'
label_path = '/content//member/seo/labels'

all_images = [f for f in os.listdir(image_path) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]
num_total = len(all_images)

# 분할 비율 0.2 0.1에 따라 이미지 몇장인지 계산
test_split = int(0.2 * num_total)
val_split = int(0.1 * num_total)
train_split = num_total - test_split - val_split
```

```python
for idx, image_file in enumerate(all_images):
    # 이미지와 라벨 폴더 경로
    src_image_path = os.path.join(image_path, image_file)
    label_file = os.path.splitext(image_file)[0] + '.txt'
    src_label_path = os.path.join(label_path, label_file)

    # train test val 에 따라 목적지 폴더 결정
    if idx < train_split:
        dest_image_path = os.path.join(train_i, image_file)
        dest_label_path = os.path.join(train_l, label_file)
    elif idx < train_split + val_split:
        dest_image_path = os.path.join(val_i, image_file)
        dest_label_path = os.path.join(val_l, label_file)
    else:
        dest_image_path = os.path.join(test_i, image_file)
        dest_label_path = os.path.join(test_l, label_file)

    # 이미지 파일
    shutil.copy(src_image_path, dest_image_path)

    # 라벨 파일이 있는 경우 동일 데이터셋으로 복사
    if os.path.exists(src_label_path):
        shutil.copy(src_label_path, dest_label_path)

print("이미지와 라벨 파일이 train, val, test 폴더에 분할되어 저장되었습니다.")
```
