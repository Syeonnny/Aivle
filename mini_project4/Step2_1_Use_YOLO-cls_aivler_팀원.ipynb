### (2) 데이터셋 전처리

* **세부 요구사항**
    - 데이터셋을 전처리 합니다.
        - YOLO-cls 모델이 요구하는 폴더 구조를 만듭니다.
            1. Datasets라는 폴더를 생성합니다.
            2. Training set, Validation set, Test set(선택 사항) 각 데이터셋이 들어갈 폴더를 생성합니다.
            3. 각 데이터셋 폴더에 분류할 클래스의 이름을 가진 폴더를 생성합니다.
        - 폴더 구조에 맞게 데이터를 분배합니다.
    - 예시 코드에서 사용한 라이브러리
        - os, glob, random, shutil, numpy

```python
import tensorflow as tf
import tensorflow_datasets as tfds
from keras.utils import image_dataset_from_directory
from sklearn.model_selection import train_test_split
import glob
import random
import shutil
import numpy as np
import pathlib
import PIL
import PIL.Image
```

#### 1) 모델이 요하는 구조의 폴더 생성

```python
path='/content'
os.makedirs(path+'/Datasets',exist_ok=True)
dataset_dir=path+'/Datasets'
train_dir = os.path.join(dataset_dir, 'train')
test_dir = os.path.join(dataset_dir, 'test')
val_dir = os.path.join(dataset_dir, 'val')

for directory in [train_dir, test_dir,val_dir]:
    os.makedirs(os.path.join(directory, 'my_face'), exist_ok=True)
    os.makedirs(os.path.join(directory, 'other_face'), exist_ok=True)

```

#### 2) 각 폴더에 이미지 데이터 옮기기

```python
my_face_images = glob.glob(path+'/my_face_ex/*.jpg')
other_face_images = glob.glob(path+'/lfw-deepfunneled/*/*.jpg')

my_face_labels = [1] * len(my_face_images)
other_face_labels = [0] * len(other_face_images)

# 이미지와 레이블을 병합하고 섞어서 학습, 검증, 테스트로 분할
all_images = my_face_images + other_face_images
all_labels = my_face_labels + other_face_labels
train_x, test_x, train_y, test_y = train_test_split(all_images, all_labels, test_size=0.2, random_state=42,stratify=all_labels)
train_x, val_x, train_y, val_y = train_test_split(train_x, train_y, test_size=0.2, random_state=42)
# 이미지 파일을 각 디렉토리로 복사
def copy_images(file_list, labels, target_dir):
    for file_path, label in zip(file_list, labels):
        class_name = 'my_face' if label == 1 else 'other_face'
        target_path = os.path.join(target_dir, class_name, os.path.basename(file_path))
        shutil.copy(file_path, target_path)


copy_images(train_x, train_y, train_dir)
copy_images(test_x, test_y, test_dir)
copy_images(val_x, val_y, val_dir)
```
